```
 import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
```

```
 from io import BytesIO
import urllib, zipfile, requests
 
r = urllib.request.urlopen('https://wri-sites.s3.amazonaws.com/climatewatch.org/www.climatewatch.org/climate-watch/climate-watch-download-zip/ghg-emissions.zip')
with zipfile.ZipFile(BytesIO(r.read())) as z:
    print( z.namelist() )
    z.extractall()
 
df = pd.read_excel('CW_CAIT_GHG_Emissions.xlsx', index_col='Country')
df
```

```
clean_df_including_LUCF = df[df.Sector == 'Total including LUCF']
clean_df_all_ghg = clean_df_including_LUCF[clean_df_including_LUCF.Gas == 'All GHG']
clean_df_all_ghg_transposed = clean_df_all_ghg.transpose()
clean_df_all_ghg_transposed = clean_df_all_ghg_transposed.drop(['Source', 'Sector','Gas'])
clean_df_all_ghg_transposed
```

```
years = clean_df_all_ghg_transposed.index.values
years = years.reshape(-1,1)
 
poly = PolynomialFeatures(degree=5)
years_poly = poly.fit_transform(years)
```

```
def predict(iso_code, name):
  values = clean_df_all_ghg_transposed[iso_code].values
  poly.fit(years_poly,values)
 
  linreg = LinearRegression()
  linreg.fit(years_poly,values)
 
  years_to_pred = years + 26
  total_years = np.concatenate((years, years_to_pred), axis=0) 
 
  years_to_pred = years_to_pred.reshape(-1,1)
  years_to_pred_poly = poly.fit_transform(years_to_pred)
  
  values_pred = np.concatenate((linreg.predict(years_poly) , linreg.predict(years_to_pred_poly)), axis=0)
 
  plt.scatter(years, values, color='blue')
  plt.plot(total_years, values_pred, color='red')
 
  plt.title(name + ' Green House Emissions')
  plt.xlabel('Year')
  plt.ylabel('GHG Emission (Mt)')
 
#  plt.savefig('/content/drive/MyDrive/' +name +' Predicted Emissions' +'.png')
  plt.show()
  return values_pred
```

```
usa = predict('USA', 'USA')
```

```
china = predict('CHN', 'China')
```

```
india = predict('IND','India')
```

```
russia = predict('RUS','Russia')
```

```
japan = predict('JPN','Japan')
```

```
germany = predict('DEU','Germany')
```

```
iran = predict('IRN','Iran')
```

```
south_korea = predict('KOR','South Korea')
```

```
saudi_arabia = predict('SAU','Saudi Arabia')
```

```
indonesia = predict('IDN','Indonesia')
```

```
 x_axis = np.concatenate((years, years + 26), axis=0)
plt.plot(x_axis,china)
plt.plot(x_axis,usa)
plt.plot(x_axis,india)
plt.plot(x_axis,russia)
plt.plot(x_axis,japan)
plt.plot(x_axis,germany)
plt.plot(x_axis,iran)
plt.plot(x_axis,south_korea)
plt.plot(x_axis,saudi_arabia)
plt.plot(x_axis,indonesia)
 
plt.legend(['China', 'USA', 'India', 'Russia', 'Japan', 'Germany', 'Iran', 'South Korea', 'Saudi Arabia', 'Indonesia'])
plt.title('Green House Gas Emissions OF All')
plt.xlabel('Year')
plt.ylabel('GHG Emissions (Mt)')
 
plt.show()
```
